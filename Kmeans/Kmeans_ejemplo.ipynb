```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples=1000, centers=4, random_state=0)
```

```python
X.shape
```

```python
X
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
```

```python
kmeans.fit(X)
```

```python
# Obtener las etiquetas de los clústeres asignados a cada punto
labels = kmeans.labels_
```

```python
labels
```

```python
# Obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
# Visualización de los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.title("K-means Clustering")
plt.show()
```

```python
# Lista para almacenar los valores de WCSS
wcss = []
```

```python
# Calcular el WCSS para diferentes valores de K
for k in range(1, 11):
 kmeans = KMeans(n_clusters=k, random_state=0)
 kmeans.fit(X)
 wcss.append(kmeans.inertia_)
```

```python
# Graficar el gráfico del codo
plt.plot(range(1, 11), wcss)
plt.xlabel('Número de clústeres (K)')
plt.ylabel('WCSS')
plt.title('Técnica del codo')
plt.show()
```

```python
import numpy as np
scores = []
range_values = range(1, 15)
for k in range_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    scores.append(kmeans.inertia_)


diff = np.diff(scores)
diff2 = np.diff(diff)
optimal_k = range_values[np.argmax(diff2) + 1]

print("Número óptimo de clusters:", optimal_k)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Visualización de los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.title("K-means Clustering")
plt.show()
```
